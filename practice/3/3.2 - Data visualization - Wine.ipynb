# Pairwise correlation of columns

Тобто відображення кореляційної матриці

```python
plt.figure(figsize=(15, 9))
corr = df.corr()#Розрахунок кореляції між всіма числовими змінними
fig, ax = plt.subplots(figsize=(5, 4)) # size of the box (розмір)
sns.heatmap(corr,  cmap='coolwarm', annot=True, fmt=".2f",center=0, square=True, cbar_kws={'label': 'Коефіцієнт кореляції'})#теплова карта матриці
plt.title('Кореляційна матриця характеристик вин', fontsize=14, pad=15)
plt.xlabel('характеристики вин', fontsize=15)
plt.ylabel('характеристики вин', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()#вивід в трохи розгорнутому вигляді
#cmap='coolwarm' - кольорова схема від синього до червоного
#annot=True - показує числові значення кореляції
#fmt=".2f" - формат чисел (2 знаки після коми)
#додав квадратні комірки та встановив центр кольорової схеми на 0, додав більше підписів
```

Парні залежності

```python
sns.pairplot(df, hue="target", palette = 'Set1')
#фарби це точки по класах вин
#Кольорова схема - червоний, синій, зелений
```

Кожен клас вина має свій колір

Матриця scatter plots з регресійними лініями

```python
plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='target', palette='bright', 
            diag_kind='hist',#Гістограми на діагоналі
            plot_kws={'alpha': 0.6},#Прозорість точок
            diag_kws={'alpha': 0.7})#Прозорість гістограм
plt.suptitle('Парні залежності з регресійними лініями', y=1.02)
plt.show()
#pairplot показує взаємозв’язки між усіма парами числових змінних у даних у вигляді точкових діаграм (scatter plots).
#на діагоналі відображаються гістограми (або KDE) розподілу окремих змінних.
```
