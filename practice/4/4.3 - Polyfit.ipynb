```python
import numpy as np #стандарт бібліотеки, візуалізація та робота з даними, масивами та математикою
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler# біліотеки для нормалізації даних
from sklearn.linear_model import SGDRegressor#бібліотека для стохастичного градієнтного спуску
from sklearn.pipeline import make_pipeline #пайплайн
```

```python
# data
N = 100# кількість точок даних
x = np.linspace(-2, 2, N)
noise = np.random.normal(0, .5, N)
y0 = x**4 - 3*x**2 + x + 5
y = y0 + noise
#Генеруэмо дані аналогічно до завдань 4.2 4.1 4.0
```

Поліноміальна регресія з polyfit

```python
# polyfit
coef = np.polyfit(x, y, deg=4)
# тут знаходимо коефіцієнти полінома 4-го степеня методом найменших квадратів
# І обчислюємо значення полінома з знайденими коефіцієнтами
y_fit = np.polyval(coef, x)
coef #  коефіцієнти c4, c3, c2, c1, c0
```

Візуалізація

```python
# display
plt.grid()
# оригінальна функція синім кольором
plt.plot(x, y0, "b", label="Original function")
plt.scatter(x, y, color="g", marker =".", label="Original + noise (data)", alpha=0.6) # зашумлені дані зеленим
plt.plot(x, y_fit, label="polyfit", color="r")#знайдена функція червоним кольором
plt.legend()
```
