```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# generation of synthetic dataset
N = 100
a = 1# коефіцієнт нахилу лінії
b = 3# вільний член (перетин з віссю Y)

x = np.random.uniform(0, 5, N)#створення випадкових значень x у діапазоні від 0 до 5
noise = np.random.normal(0, .5, N)
y = a * x + b + noise#Обчислення y згідно рівнянню прямої з шумом

plt.grid()# Побудова графіку розсіювання для ознайомлення з даними
plt.scatter(x, y, marker = '.', color = 'r')#точки на графіку
```

```python
X = x.reshape(-1, 1)#Перетворення x у двовимірний масив для sklearn
```

```python
x # 1d array
```

```python
X # array where each row is an array to
```

```python
# a * x + b = y

model = LinearRegression()# Створення лінійної регресійної моделі
model.fit(X, y)
model
```

```python
a_found = model.coef_[0]#Отримання коефіцієнтів моделі
b_found = model.intercept_
```

```python
a_found
print(f"Нахил a (коефіцієнт): {a_found:.2f}")
```

```python
b_found
print(f"Перетин b (вільний член): {b_found:.2f}")
```

Візуалізація результату лінійної регресії

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', alpha=0.6, label='Дані') #Побудова графіку розсіювання (scatter plot)
#побудова лінії регресії на основі прогнозу моделі
#model.predict(X) дає передбачені значення Y для кожного x
plt.plot(x, model.predict(X), color='blue', linewidth=2, label=f'Лінія: y = {a_found:.2f}x + {b_found:.2f}')
plt.xlabel('X')#підписи осей
plt.ylabel('Y')
plt.title('Лінійна регресія')
plt.legend()#відображення легенди графіку з підписами даних і лінії моделі
plt.grid(True)
plt.show()
#вивід у консоль знайдених параметрів регресії
print(f"Знайдено: a = {a_found:.2f}, b = {b_found:.2f}")
```
